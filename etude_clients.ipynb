```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import LinearRegression 
```

```python
df = pd.read_csv('data/clients.csv')
df 
```

```python
corr_matrix = df[['Age', 'AnnualIncome', 'PurchaseAmount', 'PurchaseFrequency', 'CustomerSatisfaction', 'Latitude', 'Longitude', 'Population', 'Altitude', 'ShippingCost']].corr(method='pearson')
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Matrice de corrélation - Pearson")
plt.show()
```

```python
y = df[['Altitude']] # Feature
X = df['PurchaseAmount'] # Target

plt.scatter(X, y)
```

```python
df = df.dropna()
```

```python
from sklearn.linear_model import LinearRegression 
model = LinearRegression()

X = df[['Altitude']]  # double crochets car dataframe nécessaire en entrée
y = df['PurchaseAmount']        # MSRP = prix conseillé

model.fit(X, y)
```

```python
slope, intercept = model.coef_[0], model.intercept_
print(f"Modèle de régression : y = {slope:.2f} * x + {intercept:.2f}")
```

```python
plt.scatter(X, y)
```

```python
# Visualisation de la droite de régression
y_pred = model.predict(X)
plt.scatter(X, y, alpha=0.7)
plt.plot(X, y_pred, color='red')
```
